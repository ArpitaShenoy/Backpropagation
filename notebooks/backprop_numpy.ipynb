```python
# forward pass
# Dummy Inputs
x = np.array([0.05, 0.10])
y = np.array([0.01,0.99])

weights1 = np.array([[0.15, 0.25],[0.20, 0.30]])
bias1 = np.array([0.35,0.35])

weights2 = np.array([[0.40, 0.5],[0.45, 0.55]])
bias2 = np.array([0.60,0.60])
```

```python
import src.LinearLayer as ll
import src.Loss as Loss

layer1 = ll.Linear(2, 2, weights=weights1, bias=bias1)
layer2 = ll.Sigmoid()
layer3 = ll.Linear(2, 2, weights=weights2, bias=bias2)
layer4 = ll.Sigmoid()

loss = Loss.MSELoss()
```

```python
# Linear function
h = np.matmul(x,weights1)+bias1

# non-linear activation(logistic)
h1 = 1/(1+np.exp(-h))

# final output layer
o = np.matmul(h1,weights2)+bias2

# non-linear activation(logistic)
o1 = 1/(1+np.exp(-o))
```

```python
o1 == layer4(layer3(layer2(layer1(x))))
```

```python
E = 0.5*(np.sum((y-o1)**2))

E == loss(layer4(layer3(layer2(layer1(x)))), y)
```

Backward propagation

```python
o1-y == loss.backward()
```

```python
# o1 w.r.t o
(o1-y)*(o1*(1-o1)) == layer4.backward(loss.backward())
```

```python
dw2 = np.matmul(h1.reshape(-1,1), ((o1-y)*(o1*(1-o1))).reshape(1,-1))
db2 = (o1-y)*(o1*(1-o1))
```

```python
np.matmul((o1-y)*(o1*(1-o1)), weights2.T) == layer3.backward(layer4.backward(loss.backward()))
```

```python
dE_dh1 = np.matmul((o1-y)*(o1*(1-o1)), weights2.T) 
```

```python
dE_dh1*(h1*(1-h1)) == layer2.backward(layer3.backward(layer4.backward(loss.backward())))
```

```python
np.matmul(dE_dh1*(h1*(1-h1)), weights1.T) == layer1.backward(layer2.backward(layer3.backward(layer4.backward(loss.backward()))))
```
